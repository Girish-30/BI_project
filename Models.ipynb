```python
gb1.Aerobic_and_Anaerobic_conditioning= gb1.Aerobic_and_Anaerobic_conditioning.astype('int') 
gb1.dtypes
```

```python
gb1['Age'] = gb1['Age'].replace({ 'Below 12yrs' : 1 , 'Below 15yrs' :2 ,'Below 18yrs' : 3, 'Above 18yrs': 4})
gb1['Improve_confidence'] = gb1['Improve_confidence'].replace({ 'Yes' : 1 , 'No' :2})
gb1['Physical_Toughness'] = gb1['Physical_Toughness'].replace({ 'Agree' : 1 , 'Disagree' :2, "Can't Say": 3})
gb1['Mental_Toughness_and_Resilience'] = gb1['Mental_Toughness_and_Resilience'].replace({ 'Strongly disagree' : 1 ,'Disagree' :2 ,'Neutral':3 ,'Agree' :4 ,'Strongly agree':5})
gb1['Fitness'] = gb1['Fitness'].replace({'Maybe':3 , 'Yes' : 1 , 'No' :2})
gb1['Overall_Body_Coordination_and_Proprioception'] = gb1['Overall_Body_Coordination_and_Proprioception'].replace({'Strongly disagree' : 1 ,'Disagree' :2 ,'Neutral':3 ,'Agree' :4 ,'Strongly Agree':5})
gb1['Stress_Relief_and_Improve_Sleep'] = gb1['Stress_Relief_and_Improve_Sleep'].replace({'Strongly disagree' : 1 ,'Disagree' :2 ,'Neutral':3 ,'Agree' :4 ,'Strongly agree':5})
gb1['Lose_weight'] = gb1['Lose_weight'].replace({'Yes' : 1 , 'No' :2 , 'May be':3})
```

```python
one_hot_data = pd.get_dummies(gb1[['Most_entertaining']])
one_hot_data
```

```python
one_hot_data = pd.get_dummies(gb1[['Most_entertaining']],drop_first=True)
one_hot_data
```

```python
merge = pd.concat([gb1,one_hot_data,],axis='columns')
merge
```

```python
last = merge.drop(['Most_entertaining'],axis='columns')
last
```

```python
last.dtypes
```

Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
num=LabelEncoder()

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Model Evaluations
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn import tree
import graphviz
from sklearn.tree import export_graphviz
```

```python
# Split data into X and y
predictors = ['Physical_Toughness','Aerobic_and_Anaerobic_conditioning']
X = last[predictors]
y = last['Lose_weight']
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Using DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Using DecisionTreeClassifier
dct=DecisionTreeClassifier(random_state=42)
```

```python
dct.fit(X_train,y_train)
```

```python
y_preds=dct.predict(X_test)
```

```python
dct.score(X_test, y_test)*100
```

```python
print(classification_report(y_test, y_preds))
```

```python
cm = confusion_matrix(y_test, y_preds)
cm
```

```python
accuracy_score(y_test, y_preds)*100
```
